```python
import pandas as pd
import datetime as dt
```

```python
# Create a DataFrame for the JC-202004-citibike-tripdata data
citibike_data = pd.read_csv(r'C:\Users\emich\Desktop\penn data science bootcamp\Module18_Assignment\Citi-Bike-Data-Tableau\Resources\JC-202004-citibike-tripdata.csv')
citibike_data.head()
```

```python
# Check datatypes of columns
citibike_data.dtypes
```

```python
# Convert the 'tripduration' column to datetime datatype.
new_tripduration = citibike_data.tripduration
# print(new_tripduration)

new_tripduration_datetime = pd.to_datetime(new_tripduration, unit='m')
# print(new_tripduration_datetime)

# Add a new column for the converted trip duration
citibike_data['new_tripduration'] = new_tripduration_datetime

# Convert the numeric gender values to string values 
citibike_data['gender'].replace([0,1,2],['Unknown','Male','Female'],inplace=True)
citibike_data.head()
```

```python
# Check the datatypes of columns. 
print(citibike_data.dtypes)
citibike_data.head()
```

```python
# Export the Dataframe as a new CSV file without the index.
citibike_data.to_csv("nyc_citi_bike_datetime_1.csv", index=False)
```
